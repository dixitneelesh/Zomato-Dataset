```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## conclusion
1. when rating is b/w 4.5 to 4.9 ---> Excellent
2. when rating b/w 4.0 to 4.4 -----> very good
3. when rating b/w 3.5 to 3.9 -------> good
4. when rating b/w 3.0 to 3.4 -----> Average
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y= 'Rating Count', data= ratings)
```

```python
sns.barplot(x='Aggregate rating', y= 'Rating Count',hue='Rating color', data= ratings, palette=['white', 'red', 'orange','green', 'green'])
```

```python
sns.countplot(x='Rating color', data = ratings, palette=['white', 'red', 'orange','green', 'green'])
```
